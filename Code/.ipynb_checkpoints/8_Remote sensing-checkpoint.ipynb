## Zonal statistics
Our datasets are cleaned and ready to be use. So, what shall we do now? The most interesting part is to compute zonal statistics. We are just computing the average of each data within a specified spatial area. 

Lets do it at the *woreda* (district level) in Ethiopia.

```python
popd  = popd_cln
suit = suit_cln

#Make a list of rasters that we wish to aggregate
RASTS = [popd_cln, suit_cln]
#Matching list of names
PREFS = ['pd',  'ag_suit']
```

```python
woreda_gdf = woreda_gdf[['OBJECTID', 'geometry', 'WOR_P_CODE']] #keep only necessary variables in woreda shapefile
```

**Zonal statistics **

The following piece of code gives us the mean for each raster

```python
start = time.time() #let's time how long each loop takes 
for idx, rast in enumerate(RASTS): # here we loop over the rasters some zonal statistics
    pref = PREFS[idx] #get our preferred name of the raster
    stat_list=['mean'] #set list of stats to calculate, for now just do the mean
    zstats=rasterstats.zonal_stats(woreda_gdf, rast, stats=stat_list) #use the zonal stat package on all raster listed and put the stat in stat list
    for st in stat_list: #add all stats to dataframe
        col_name=pref+st
        col_stat=[d[st] for d in zstats] #zonalstats outputs a list of dictionaries, this creates a list of values for stat
        woreda_gdf[col_name]=col_stat #create a new column that would be equal to the mean of that raster
    time_elapsed = round(time.time()-start)
    print('It took {dt} seconds to run zonal stats (mean) for {p}'.format(dt=time_elapsed,p=pref))
    start = time.time()
```

Nice to see that we fixed the projection. The next steps to do such calculation is to set up a starting point within each district. To do so, we will extract the `centroid` of each district. The centroid is just the center of a geometry.

We're ready to calculate the distance between each centroid and the nearest zone, regional capital, and Addis Ababa. 
1. Creates a new column equals to the nearest distance. 
2. apply() = apply a function to each row (x) for any single row x
3. Take the town and calculate the distance with each row (centroid) and find the minimum. 
4. Then divid it by 1000 to obtain the km.  

We get back to WGS to have the coordinates in degrees because equal area projection does not keep the distance equally.

```python
woreda_centroid = woreda_centroid.to_crs(crs=crs_p) #reproject to WGS
addis_gdf = addis_gdf.to_crs(crs=crs_p) 
region_gdf = region_gdf.to_crs(crs=crs_p) 
zone_gdf = zone_gdf.to_crs(crs=crs_p) 
```

We define the following function to calculate the distance between the centroid and the nearest object.

```python
#take the lon and lat of pair of coordinate and calculate the distance on the earth
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """

    lon1, lat1, lon2, lat2 = map(numpy.radians, [lon1, lat1, lon2, lat2]) # convert decimal degrees to radians 

    # haversine formula 
    dlon,dlat = lon2 - lon1 , lat2 - lat1 
    a = numpy.sin(dlat/2)**2 + numpy.cos(lat1) * numpy.cos(lat2) * numpy.sin(dlon/2)**2
    c = 2 * numpy.arcsin(numpy.sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles.
    return c * r
```

1. Find for any single woreda the nearest Addis, Zonal, and Regional capital and calculate the distance between the centroid and it. 
2. Take the geometry column (centroid) and apply this function for each row (x): the nearest point which is a shapely function. 
3. Take the nearest point from the centroid to the point. It returns a list of points: keep only the first element [0] wich is the nearest point.

```python
region_pt=region_gdf.dissolve() #take the point df and dissolve it (geo df with one row with all coordinates of the town 
addis_pt=addis_gdf.dissolve() #take the point df and dissolve it (geo df with one row with all coordinates of the town 
zone_pt=zone_gdf.dissolve() #take the point df and dissolve it (geo df with one row with all coordinates of the town 
```

```python
woreda_centroid['reg_coords'] = woreda_centroid.apply(lambda x: nearest_points(region_pt.geometry[0],x['geometry'])[0], axis=1)
woreda_centroid['addis_coords'] = woreda_centroid.apply(lambda x: nearest_points(addis_pt.geometry[0],x['geometry'])[0], axis=1)
woreda_centroid['zone_coords'] = woreda_centroid.apply(lambda x: nearest_points(zone_pt.geometry[0],x['geometry'])[0], axis=1)
```

We have found for each the nearest regional and zonal capital, as well as Addis Ababa. We can add the coordinates of this nearest point to the existing dataset.

```python
#create 8 different columns with lat and lon of the pair of coordinates
woreda_centroid['lon']=woreda_centroid.geometry.x 
woreda_centroid['lat']=woreda_centroid.geometry.y
woreda_centroid['lon_addis']=woreda_centroid.addis_coords.apply(lambda p: p.x)
woreda_centroid['lat_addis']=woreda_centroid.addis_coords.apply(lambda p: p.y)
woreda_centroid['lon_zone']=woreda_centroid.zone_coords.apply(lambda p: p.x)
woreda_centroid['lat_zone']=woreda_centroid.zone_coords.apply(lambda p: p.y)
woreda_centroid['lon_reg']=woreda_centroid.reg_coords.apply(lambda p: p.x)
woreda_centroid['lat_reg']=woreda_centroid.reg_coords.apply(lambda p: p.y)
```

Use the formula we created before to calculate the distance.


```python
woreda_centroid['dist2addis_ea']=haversine(woreda_centroid['lon'],woreda_centroid['lat'],woreda_centroid['lon_addis'],woreda_centroid['lat_addis']) #calculate nearest distance to addis for each woreda centroid
woreda_centroid['dist2reg_ea']=haversine(woreda_centroid['lon'],woreda_centroid['lat'],woreda_centroid['lon_reg'],woreda_centroid['lat_reg']) #calculate nearest distance to reg capital for each woreda centroid
woreda_centroid['dist2zone_ea']=haversine(woreda_centroid['lon'],woreda_centroid['lat'],woreda_centroid['lon_zone'],woreda_centroid['lat_zone']) #calculate nearest distance to zone capitla for each woreda centroid
```
